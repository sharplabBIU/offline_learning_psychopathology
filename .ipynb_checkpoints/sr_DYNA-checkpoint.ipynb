# SR DYNA Simulation: Replicating Russek Et al

```python
import imageio
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import softmax
from random import sample as sampr

class SRDYNA:
    def __init__(self, maze_size, gamma,epsilon, alpha_sr, alpha_td, k, reward_location,punishment_location, obstacles,reward_value,punishment_value, dynamic_lr=False, ep_num=1,total_eps=10,rest_episode=[],specific_trial=1):
        #features of Maze environment
        self.maze_size = maze_size
        self.obstacles = obstacles # hitting obstacle is a special case
        self.reward_location = reward_location 
        self.reward_value=reward_value
        self.punishment_value=punishment_value
        self.punishment_location = punishment_location
        self.num_states = maze_size * maze_size
        self.num_actions = 4 # Up, Down, Left, Right
        
        #parameters of learning
        self.gamma = gamma
        self.epsilon=epsilon
        self.alpha_sr = alpha_sr
        self.alpha_td = alpha_td

        #if you want to engage offline sampling at specific points
        self.rest_episode=rest_episode
        self.specific_trial=specific_trial

        #number of samples each offline DYNA update
        self.k = k

        
        #keeping track of data
        self.episode = ep_num
        self.episodes=range(total_eps)
        self.current_trial = 0

        #SR matrix (sa,s'a') form
        self.H = np.zeros((self.num_states * self.num_actions, self.num_states * self.num_actions))

        #Q values computed through planning (i.e., dot_product(H,W))
        self.Q = np.zeros((self.num_states*self.num_actions))

        #Learned reward weights
        self.W = np.zeros((self.num_states * self.num_actions,))

        # for dynamic LR 
        self.dynamic_lr = dynamic_lr
        self.state_action_counts = np.ones((self.num_states, self.num_actions))
        
        #performance measure
        self.episode_goal_counts = np.zeros((total_eps))

        #remember
        self.dn_tot = 0
        self.sa = [{'samples': np.zeros((1000000, 4)), 'dn': 0} for _ in range(self.num_states * self.num_actions)] 
        self.sa_list = np.zeros(1000000)

        
    def state_action_to_index(self, state, action):
        return state * self.num_actions + action

    def index_to_state_action(self, index):
        return divmod(index, self.num_actions)

    def update_sr_online(self, state, action, next_state, next_action,phase):
        if state == next_state:
            return

        sa_index = self.state_action_to_index(state, action)
        # print('sa index: {}, state: {}'.format(sa_index,state))
        next_sa_index = self.state_action_to_index(next_state, next_action)

        identity_sa = np.zeros((self.num_states * self.num_actions,))
        identity_sa[sa_index] = 1

        if self.dynamic_lr:
            
            
            alpha_sr = 1.0 / self.state_action_counts[state, action]
            self.state_action_counts[state, action] += 1
            
        else:
            alpha_sr = self.alpha_sr

        self.H[sa_index, :] += alpha_sr * (identity_sa + self.gamma * self.H[next_sa_index, :] - self.H[sa_index, :])
        
        self.sa[sa_index]['samples'][self.sa[sa_index]['dn']] = [state, action, next_state, next_action] 
        self.sa[sa_index]['dn'] += 1 
        self.sa_list[self.dn_tot] = sa_index 
        self.dn_tot += 1
    

        
                
        
    def update_weights(self, state, action, reward, next_state, next_action):

        sa_index = self.state_action_to_index(state, action)
        next_sa_index = self.state_action_to_index(state, action)
        
        PE = reward + self.gamma * self.Q[next_sa_index] - self.Q[sa_index]
        norm = np.linalg.norm(self.H[sa_index, :])
        if norm > 1e-6:
            feature_rep = self.H[sa_index, :] / norm
        else:
            feature_rep = self.H[sa_index, :]
 
                
        self.W += self.alpha_td * PE * feature_rep
                
        
    def plot_sr_matrix(self):
        plt.figure(figsize=(10, 8))
        plt.imshow(self.H, cmap='viridis', aspect='auto')
        plt.colorbar()
        plt.title('Successor Representation (SR) Matrix H')
        plt.xlabel('State-Action Index')
        plt.ylabel('State-Action Index')
        plt.show()


    def update_sr_offline(self):
        u_sa = np.unique(self.sa_list[:self.dn_tot])

        for simul in range(self.k):
            try:
                ind = np.random.randint(len(u_sa))
            except:
                print(u_sa)
                np.random.randint(len(u_sa))
            sa_num = u_sa[ind]
            sa_num=int(sa_num)
            
            while self.sa[sa_num]['dn']<1:
                ind = np.random.randint(len(u_sa))
                sa_num = u_sa[ind]
                sa_num=int(sa_num)
            if self.sa[sa_num]['dn']>50:
                pdf = np.exp(np.arange(self.sa[sa_num]['dn']-50, self.sa[sa_num]['dn']))
                mem_length=50
            else:
                pdf = np.exp(np.arange(self.sa[sa_num]['dn']))
                mem_length=self.sa[sa_num]['dn']
            
            pdf /= pdf.sum()
            rd=np.random.rand()
            n = np.where( rd< np.cumsum(pdf))[0][0]
            
            if n > self.sa[sa_num]['dn'] - 1:
                n = self.sa[sa_num]['dn'] - 1
         
            
            sample = self.sa[sa_num]['samples'][self.sa[sa_num]['dn']-mem_length+n]
            
            state, action, next_state, next_action = sample
            state=int(state)
            action=int(action)
            next_state=int(next_state)
            next_action=int(next_action)
            sa_index = int(self.state_action_to_index(state, action))
            next_sa_index = int(self.state_action_to_index(next_state, next_action))
            
            identity_sa = np.zeros((self.num_states * self.num_actions,))
            identity_sa[sa_index] = 1
    
            if self.dynamic_lr:                
                alpha_sr = 1.0 / self.state_action_counts[state, action]
                # self.state_action_counts[state, action] += 1
            else:
                alpha_sr = self.alpha_sr
            sa_max_poss = int(self.state_action_to_index(next_state, 0))
            Q_vals=[]
            for maxl in range(self.num_actions):
                Q_vals.append(self.Q[sa_max_poss+maxl])
             
            maxA=Q_vals.index(max(Q_vals))
            
            max_sa_index = int(self.state_action_to_index(next_state, maxA))

            self.H[sa_index, :] += alpha_sr * (identity_sa + self.gamma * self.H[max_sa_index, :] - self.H[sa_index, :])
            
            self.Q=np.matmul(self.H,self.W)


    def modelchoose(self, Qvals, epsilon):
        action_prob = np.zeros(len(Qvals))
        maxval = np.max(Qvals)        
        bestA = np.where(Qvals == maxval)[0]
        action_prob[bestA] = (1 - epsilon) / len(bestA)
        action_prob += epsilon / len(Qvals)
        action = np.where(np.random.rand() < np.cumsum(action_prob))[0][0]
        return action

    def plan_q_policy(self, state, epsilon):
        # Identify valid actions
        valid_actions = []
        valid_sa_indices = []
        action_to_original_index = {}  # Map valid action index to its original index
        for action in range(self.num_actions):
            next_state = self.transition_function(state, action)
            if next_state != state:  # Only consider actions leading to valid transitions
                valid_actions.append(action)
                valid_sa_indices.append(self.state_action_to_index(state, action))
                action_to_original_index[len(valid_actions) - 1] = action  # Map valid action index
        
        # If no valid actions are available (edge case), return a random valid action
        if not valid_actions:
            return random.choice(range(self.num_actions))
        
        # Compute Q-values for valid actions
        q_values = [np.dot(self.H[sa], self.W) for sa in valid_sa_indices]
        # print('plan for state:{}, q_vals:{}'.format(state,q_values))
        
        # Use epsilon-greedy policy to select an action index among valid actions
        chosen_valid_index = self.modelchoose(q_values, epsilon)
        
        # Map back to the original action index
        chosen_action = action_to_original_index[chosen_valid_index]
        
        # Update Q-values for visualization or logging
        self.Q = np.matmul(self.H, self.W)
        
        return chosen_action

    

    def transition_function(self, state, action):
        if not isinstance(state, int):
            raise ValueError(f"Invalid state type: {type(state)}. State must be an integer.")
        row, col = divmod(state, self.maze_size)
        
        if action == 0 and row > 0:  # Up
            next_state = (row - 1) * self.maze_size + col
        elif action == 1 and col > 0:  # Left
            next_state = row * self.maze_size + (col - 1)
        elif action == 2 and col < self.maze_size - 1:  # Right
            next_state = row * self.maze_size + (col + 1)
        elif action == 3 and row < self.maze_size - 1:  # Down
            next_state = (row + 1) * self.maze_size + col
        else:
            next_state = state
    
        if next_state in [ob[0] * self.maze_size + ob[1] for ob in self.obstacles]:
            next_state = state
    
        return next_state


    def step(self, state, action, reward, next_state, next_action, phase='warmup'):
        # if state!=next_state:
        self.update_sr_online(state, action, next_state, next_action, phase)
        self.update_weights(state, action, reward, next_state, next_action)
        if phase == 'offline':  # Engage offline learning only during the specified phase
            self.update_sr_offline()
        self.current_trial += 1

    def run_episode(self, ep_num, max_steps=1000,phase_general='warmup'):
        start_state = 0
        state = start_state
        trajectory = []  # Record the trajectory
        self.episode = ep_num
        self.current_trial = 0
        reached_goal_trial = 0
        if self.episode ==5:
            self.obstacles = [(1, 0), (2, 0), (2, 1)]
        for _ in range(max_steps):
            action = self.plan_q_policy(state, self.epsilon)
            next_state = self.transition_function(state, action)
           
            trajectory.append((state, action))  # Record state and action
            if phase_general!='warmup':
                reward = self.reward_value if (next_state == self.reward_location[0] * self.maze_size + self.reward_location[1]) else (self.punishment_value if (next_state == self.punishment_location[0] * self.maze_size + self.punishment_location[1]) else -1)
            else:
                reward=-1
            next_action = self.plan_q_policy(next_state, self.epsilon)
            phase = "offline" if self.rest_episode > 0 else "choice"
            self.step(state, action, reward, next_state, next_action, phase)
            if reward == self.reward_value:
                reached_goal_trial = self.current_trial
                self.episode_goal_counts[self.episode] += 1
                state=0
                
            else:
                state = next_state
        return trajectory  # Return the trajectory

            

            
      

    def plot_goal_counts(self, goal_counts, agent_labels):
        plt.figure(figsize=(10, 6))
        for goals, label in zip(goal_counts, agent_labels):
            
            mean_goals = np.mean(goals, axis=0)
            std_error = np.std(goals, axis=0) / np.sqrt(goals.shape[0])
            plt.plot(self.episodes, mean_goals, label=label)
            plt.fill_between(self.episodes, mean_goals - std_error, mean_goals + std_error, alpha=0.2)
        plt.xlabel('Episode')
        plt.ylabel('Number of times goal was reached')
        plt.title('Goal counts per episode')
        plt.legend()
        plt.savefig('SRDyna_Sim_DynamicLR.png',dpi=300,bbox_inches='tight')
        plt.show()
        
    def plot_goal_counts_single(self):
        ax=sns.barplot(x=self.episodes,y=self.episode_goal_counts)
        plt.xlabel('Episode')
        plt.ylabel('Number of times goal was reached')
        plt.title('Goal counts per episode')
        plt.legend()
        plt.show()
        
    def plot_sr_matrix(self):
        plt.figure(figsize=(10, 8))
        plt.imshow(self.H, cmap='viridis', aspect='auto')
        plt.colorbar()
        plt.title('Successor Representation (SR) Matrix H')
        plt.xlabel('State-Action Index')
        plt.ylabel('State-Action Index')
        plt.show()

    def plot_W_vector(self):
        sns.barplot(y=self.W,x=np.arange(1,len(self.W)+1))
        plt.xticks([0,50,100,150])
        plt.title('Reward Function')
        plt.show()

    def plot_maze_with_agent(self, agent_state):
        maze = np.zeros((self.maze_size, self.maze_size, 3))
        maze[:] = [0.5, 0.5, 0.5]
        maze[self.reward_location[0], self.reward_location[1]] = [1.0, 0.84, 0.0]
        maze[self.punishment_location[0], self.punishment_location[1]] = [1.0, 0.0, 0.0]
        for ob in self.obstacles:
            maze[ob[0], ob[1]] = [0.0, 0.0, 0.0]
        start_state = 0
        start_row, start_col = divmod(start_state, self.maze_size)
        maze[start_row, start_col] = [0.0, 0.0, 1.0]
        agent_row, agent_col = divmod(agent_state, self.maze_size)
        maze[agent_row, agent_col] = [0.0, 1.0, 0.0]
        
        plt.figure(figsize=(6, 6))
        plt.imshow(maze, interpolation='none')
        plt.text(start_col, start_row, 'START', ha='center', va='center', color='white', fontsize=12, fontweight='bold')
        plt.title('Maze with Agent')
        plt.close()
        return maze


def visualize_last_episode_trajectory(agent, trajectory, output_filename='last_episode.gif'):
    """
    Visualize the agent's trajectory in the last episode.
    Generates a GIF showing the agent's movements on the maze.
    
    Parameters:
        agent (SRDYNA): The SRDYNA agent.
        trajectory (list): List of (state, action) tuples representing the trajectory.
        output_filename (str): Filename for the output GIF.
    """
    frames = []
    for state, action in trajectory:
        maze_image = agent.plot_maze_with_agent(state)
        plt.imshow(maze_image, interpolation='none')
        plt.title(f'State: {state}, Action: {["Up", "Left", "Right", "Down"][action]}')
        plt.axis('off')
        # Save frame as an image
        frame_filename = f'temp_frame_{state}_{action}.png'
        plt.savefig(frame_filename)
        frames.append(imageio.imread(frame_filename))
        plt.close()

    # Save all frames as a GIF
    imageio.mimsave(output_filename, frames, fps=2)
    print(f'Last episode trajectory saved as {output_filename}')

# Updated run function to handle specific offline trial engagement
def run_agent_with_specific_offline_trials(alpha_td=0.5, alpha_sr=0.5, offline_samples=0, trials_per_episode=100, episodes=5, specific_trial=10, dynamic_lr=True,warmup_trials=0,rest_episode=0,reward_trials=0,epsilon=0.10,gamma=0.7):
    
    maze_size = 6
    reward_value = 100
    punishment_value = -1
    reward_location = (3, 1)
    punishment_location = (1, 2)
    obstacles = []
    
    # Initialize the agent
    agent = SRDYNA(maze_size, gamma, epsilon, alpha_sr, alpha_td, offline_samples, 
                   reward_location, punishment_location, obstacles, reward_value, punishment_value, 
                   dynamic_lr, total_eps=episodes, rest_episode=rest_episode,specific_trial=specific_trial)
    
    # Warm-up Phase: Random Exploration -1000 trials default
    trajectory=agent.run_episode(0, max_steps=warmup_trials,phase_general='warmup')

    for ep in range(episodes):
        trajectory=agent.run_episode(ep, max_steps=trials_per_episode,phase_general='choice')
        
    return agent, agent.episode_goal_counts


# Configuration for multiple offline sample settings
specific_episodes = [5,-1]  # Offline samples from 10 to 1000 in steps of 100
num_agents_per_setting = 10 # 10 agents per setting
trials_per_ep = 65
eps = 10
alpha_td = 0.2
alpha_sr = 0.2
epsilon=0.05
offline_samples=1000


warmups=5000
gamma=0.99
# Collect data for all settings


data = []
for specific_episode in specific_episodes:
    print(specific_episode)
    for agent_id in range(num_agents_per_setting):
        agent, goal_counts = run_agent_with_specific_offline_trials(
            alpha_td=alpha_td,
            alpha_sr=alpha_sr,
            offline_samples=offline_samples,
            trials_per_episode=trials_per_ep,
            episodes=eps,
            specific_trial=2,
            dynamic_lr=False,
            warmup_trials=warmups,
            rest_episode=specific_episode,
            reward_trials=reward_trials,
            epsilon=epsilon,
            gamma=gamma
        )
        for episode_idx, count in enumerate(goal_counts):
            data.append({
                'OfflineStart': specific_episode,
                'Agent': f'Agent_{offline_samples}_{agent_id}',
                'Episode': episode_idx,
                'Goal Count': count
            })

```

```python
# Convert to DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Plotting with seaborn
plt.figure(figsize=(12, 8))
sns.set_palette("Set2")

sns.lineplot(
    data=df,
    x='Episode',
    y='Goal Count',
    hue='OfflineStart',
    ci=68,  # 68% confidence interval
)

plt.xlabel('Episode')
plt.axvline(x=5,ymax=18,color='red')
plt.ylabel('Number of times goal was reached')
plt.title('Goal Counts per Episode by Offline Sample Settings')
plt.legend(title='Offline Episode Initiation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('effect_offline_timing_on_performance.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
import imageio
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import softmax
from random import sample as sampr

class SRDYNA:
    def __init__(self, maze_size, gamma,epsilon, alpha_sr, alpha_td, k, reward_location,punishment_location, obstacles,reward_value,punishment_value, dynamic_lr=False, ep_num=1,total_eps=10,rest_episode=[],specific_trial=1):
        #features of Maze environment
        self.maze_size = maze_size
        self.obstacles = obstacles # hitting obstacle is a special case
        self.reward_location = reward_location 
        self.reward_value=reward_value
        self.punishment_value=punishment_value
        self.punishment_location = punishment_location
        self.num_states = maze_size * maze_size
        self.num_actions = 4 # Up, Down, Left, Right
        
        #parameters of learning
        self.gamma = gamma
        self.epsilon=epsilon
        self.alpha_sr = alpha_sr
        self.alpha_td = alpha_td

        #if you want to engage offline sampling at specific points
        self.rest_episode=rest_episode
        self.specific_trial=specific_trial

        #number of samples each offline DYNA update
        self.k = k

        
        #keeping track of data
        self.episode = ep_num
        self.episodes=range(total_eps)
        self.current_trial = 0

        #SR matrix (sa,s'a') form
        self.H = np.zeros((self.num_states * self.num_actions, self.num_states * self.num_actions))

        #Q values computed through planning (i.e., dot_product(H,W))
        self.Q = np.zeros((self.num_states*self.num_actions))

        #Learned reward weights
        self.W = np.zeros((self.num_states * self.num_actions,))

        # for dynamic LR 
        self.dynamic_lr = dynamic_lr
        self.state_action_counts = np.ones((self.num_states, self.num_actions))
        
        #performance measure
        self.episode_goal_counts = np.zeros((total_eps))

        #remember
        self.dn_tot = 0
        self.sa = [{'samples': np.zeros((1000000, 4)), 'dn': 0} for _ in range(self.num_states * self.num_actions)] 
        self.sa_list = np.zeros(1000000)

        
    def state_action_to_index(self, state, action):
        return state * self.num_actions + action

    def index_to_state_action(self, index):
        return divmod(index, self.num_actions)

    def update_sr_online(self, state, action, next_state, next_action,phase):
        if state == next_state:
            return

        sa_index = self.state_action_to_index(state, action)
        # print('sa index: {}, state: {}'.format(sa_index,state))
        next_sa_index = self.state_action_to_index(next_state, next_action)

        identity_sa = np.zeros((self.num_states * self.num_actions,))
        identity_sa[sa_index] = 1

        if self.dynamic_lr:
            
            
            alpha_sr = 1.0 / self.state_action_counts[state, action]
            self.state_action_counts[state, action] += 1
            
        else:
            alpha_sr = self.alpha_sr

        self.H[sa_index, :] += alpha_sr * (identity_sa + self.gamma * self.H[next_sa_index, :] - self.H[sa_index, :])
        
        self.sa[sa_index]['samples'][self.sa[sa_index]['dn']] = [state, action, next_state, next_action] 
        self.sa[sa_index]['dn'] += 1 
        self.sa_list[self.dn_tot] = sa_index 
        self.dn_tot += 1
    

        
                
        
    def update_weights(self, state, action, reward, next_state, next_action):

        sa_index = self.state_action_to_index(state, action)
        next_sa_index = self.state_action_to_index(state, action)
        
        PE = reward + self.gamma * self.Q[next_sa_index] - self.Q[sa_index]
        norm = np.linalg.norm(self.H[sa_index, :])
        if norm > 1e-6:
            feature_rep = self.H[sa_index, :] / norm
        else:
            feature_rep = self.H[sa_index, :]
 
                
        self.W += self.alpha_td * PE * feature_rep
                
        
    def plot_sr_matrix(self):
        plt.figure(figsize=(10, 8))
        plt.imshow(self.H, cmap='viridis', aspect='auto')
        plt.colorbar()
        plt.title('Successor Representation (SR) Matrix H')
        plt.xlabel('State-Action Index')
        plt.ylabel('State-Action Index')
        plt.show()


    def update_sr_offline(self):
        u_sa = np.unique(self.sa_list[:self.dn_tot])

        for simul in range(self.k):
            try:
                ind = np.random.randint(len(u_sa))
            except:
                print(u_sa)
                np.random.randint(len(u_sa))
            sa_num = u_sa[ind]
            sa_num=int(sa_num)
            
            while self.sa[sa_num]['dn']<1:
                ind = np.random.randint(len(u_sa))
                sa_num = u_sa[ind]
                sa_num=int(sa_num)
            if self.sa[sa_num]['dn']>50:
                pdf = np.exp(np.arange(self.sa[sa_num]['dn']-50, self.sa[sa_num]['dn']))
                mem_length=50
            else:
                pdf = np.exp(np.arange(self.sa[sa_num]['dn']))
                mem_length=self.sa[sa_num]['dn']
            
            pdf /= pdf.sum()
            rd=np.random.rand()
            n = np.where( rd< np.cumsum(pdf))[0][0]
            
            if n > self.sa[sa_num]['dn'] - 1:
                n = self.sa[sa_num]['dn'] - 1
         
            
            sample = self.sa[sa_num]['samples'][self.sa[sa_num]['dn']-mem_length+n]
            
            state, action, next_state, next_action = sample
            state=int(state)
            action=int(action)
            next_state=int(next_state)
            next_action=int(next_action)
            sa_index = int(self.state_action_to_index(state, action))
            next_sa_index = int(self.state_action_to_index(next_state, next_action))
            
            identity_sa = np.zeros((self.num_states * self.num_actions,))
            identity_sa[sa_index] = 1
    
            if self.dynamic_lr:                
                alpha_sr = 1.0 / self.state_action_counts[state, action]
                # self.state_action_counts[state, action] += 1
            else:
                alpha_sr = self.alpha_sr
            sa_max_poss = int(self.state_action_to_index(next_state, 0))
            Q_vals=[]
            for maxl in range(self.num_actions):
                Q_vals.append(self.Q[sa_max_poss+maxl])
             
            maxA=Q_vals.index(max(Q_vals))
            
            max_sa_index = int(self.state_action_to_index(next_state, maxA))

            self.H[sa_index, :] += alpha_sr * (identity_sa + self.gamma * self.H[max_sa_index, :] - self.H[sa_index, :])
            
            self.Q=np.matmul(self.H,self.W)


    def modelchoose(self, Qvals, epsilon):
        action_prob = np.zeros(len(Qvals))
        maxval = np.max(Qvals)        
        bestA = np.where(Qvals == maxval)[0]
        action_prob[bestA] = (1 - epsilon) / len(bestA)
        action_prob += epsilon / len(Qvals)
        action = np.where(np.random.rand() < np.cumsum(action_prob))[0][0]
        return action

    def plan_q_policy(self, state, epsilon):
        # Identify valid actions
        valid_actions = []
        valid_sa_indices = []
        action_to_original_index = {}  # Map valid action index to its original index
        for action in range(self.num_actions):
            next_state = self.transition_function(state, action)
            if next_state != state:  # Only consider actions leading to valid transitions
                valid_actions.append(action)
                valid_sa_indices.append(self.state_action_to_index(state, action))
                action_to_original_index[len(valid_actions) - 1] = action  # Map valid action index
        
        # If no valid actions are available (edge case), return a random valid action
        if not valid_actions:
            return random.choice(range(self.num_actions))
        
        # Compute Q-values for valid actions
        q_values = [np.dot(self.H[sa], self.W) for sa in valid_sa_indices]
        # print('plan for state:{}, q_vals:{}'.format(state,q_values))
        
        # Use epsilon-greedy policy to select an action index among valid actions
        chosen_valid_index = self.modelchoose(q_values, epsilon)
        
        # Map back to the original action index
        chosen_action = action_to_original_index[chosen_valid_index]
        
        # Update Q-values for visualization or logging
        self.Q = np.matmul(self.H, self.W)
        
        return chosen_action

    

    def transition_function(self, state, action):
        if not isinstance(state, int):
            raise ValueError(f"Invalid state type: {type(state)}. State must be an integer.")
        row, col = divmod(state, self.maze_size)
        
        if action == 0 and row > 0:  # Up
            next_state = (row - 1) * self.maze_size + col
        elif action == 1 and col > 0:  # Left
            next_state = row * self.maze_size + (col - 1)
        elif action == 2 and col < self.maze_size - 1:  # Right
            next_state = row * self.maze_size + (col + 1)
        elif action == 3 and row < self.maze_size - 1:  # Down
            next_state = (row + 1) * self.maze_size + col
        else:
            next_state = state
    
        if next_state in [ob[0] * self.maze_size + ob[1] for ob in self.obstacles]:
            next_state = state
    
        return next_state


    def step(self, state, action, reward, next_state, next_action, phase='warmup'):
        # if state!=next_state:
        self.update_sr_online(state, action, next_state, next_action, phase)
        self.update_weights(state, action, reward, next_state, next_action)
        if phase == 'offline':  # Engage offline learning only during the specified phase
            self.update_sr_offline()
        self.current_trial += 1

    def run_episode(self, ep_num, max_steps=1000,phase_general='warmup'):
        start_state = 0
        state = start_state
        trajectory = []  # Record the trajectory
        self.episode = ep_num
        self.current_trial = 0
        reached_goal_trial = 0
        if self.episode ==5:
            self.obstacles = [(1, 0), (2, 0), (2, 1)]
        for _ in range(max_steps):
            action = self.plan_q_policy(state, self.epsilon)
            next_state = self.transition_function(state, action)
           
            trajectory.append((state, action))  # Record state and action
            if phase_general!='warmup':
                reward = self.reward_value if (next_state == self.reward_location[0] * self.maze_size + self.reward_location[1]) else (self.punishment_value if (next_state == self.punishment_location[0] * self.maze_size + self.punishment_location[1]) else -1)
                phase='offline' if self.rest_episode == self.episode else "choice"
            else:
                reward=-1
                phase='warmup'
            next_action = self.plan_q_policy(next_state, self.epsilon)
            
            self.step(state, action, reward, next_state, next_action, phase)
            if reward == self.reward_value:
                reached_goal_trial = self.current_trial
                self.episode_goal_counts[self.episode] += 1
                state=0
                
            else:
                state = next_state
        return trajectory  # Return the trajectory

            

            
      

    def plot_goal_counts(self, goal_counts, agent_labels):
        plt.figure(figsize=(10, 6))
        for goals, label in zip(goal_counts, agent_labels):
            
            mean_goals = np.mean(goals, axis=0)
            std_error = np.std(goals, axis=0) / np.sqrt(goals.shape[0])
            plt.plot(self.episodes, mean_goals, label=label)
            plt.fill_between(self.episodes, mean_goals - std_error, mean_goals + std_error, alpha=0.2)
        plt.xlabel('Episode')
        plt.ylabel('Number of times goal was reached')
        plt.title('Goal counts per episode')
        plt.legend()
        plt.savefig('SRDyna_Sim_DynamicLR.png',dpi=300,bbox_inches='tight')
        plt.show()
        
    def plot_goal_counts_single(self):
        ax=sns.barplot(x=self.episodes,y=self.episode_goal_counts)
        plt.xlabel('Episode')
        plt.ylabel('Number of times goal was reached')
        plt.title('Goal counts per episode')
        plt.legend()
        plt.show()
        
    def plot_sr_matrix(self):
        plt.figure(figsize=(10, 8))
        plt.imshow(self.H, cmap='viridis', aspect='auto')
        plt.colorbar()
        plt.title('Successor Representation (SR) Matrix H')
        plt.xlabel('State-Action Index')
        plt.ylabel('State-Action Index')
        plt.show()

    def plot_W_vector(self):
        sns.barplot(y=self.W,x=np.arange(1,len(self.W)+1))
        plt.xticks([0,50,100,150])
        plt.title('Reward Function')
        plt.show()

    def plot_maze_with_agent(self, agent_state):
        maze = np.zeros((self.maze_size, self.maze_size, 3))
        maze[:] = [0.5, 0.5, 0.5]
        maze[self.reward_location[0], self.reward_location[1]] = [1.0, 0.84, 0.0]
        maze[self.punishment_location[0], self.punishment_location[1]] = [1.0, 0.0, 0.0]
        for ob in self.obstacles:
            maze[ob[0], ob[1]] = [0.0, 0.0, 0.0]
        start_state = 0
        start_row, start_col = divmod(start_state, self.maze_size)
        maze[start_row, start_col] = [0.0, 0.0, 1.0]
        agent_row, agent_col = divmod(agent_state, self.maze_size)
        maze[agent_row, agent_col] = [0.0, 1.0, 0.0]
        
        plt.figure(figsize=(6, 6))
        plt.imshow(maze, interpolation='none')
        plt.text(start_col, start_row, 'START', ha='center', va='center', color='white', fontsize=12, fontweight='bold')
        plt.title('Maze with Agent')
        plt.close()
        return maze


def visualize_last_episode_trajectory(agent, trajectory, output_filename='last_episode.gif'):
    """
    Visualize the agent's trajectory in the last episode.
    Generates a GIF showing the agent's movements on the maze.
    
    Parameters:
        agent (SRDYNA): The SRDYNA agent.
        trajectory (list): List of (state, action) tuples representing the trajectory.
        output_filename (str): Filename for the output GIF.
    """
    frames = []
    for state, action in trajectory:
        maze_image = agent.plot_maze_with_agent(state)
        plt.imshow(maze_image, interpolation='none')
        plt.title(f'State: {state}, Action: {["Up", "Left", "Right", "Down"][action]}')
        plt.axis('off')
        # Save frame as an image
        frame_filename = f'temp_frame_{state}_{action}.png'
        plt.savefig(frame_filename)
        frames.append(imageio.imread(frame_filename))
        plt.close()

    # Save all frames as a GIF
    imageio.mimsave(output_filename, frames, fps=2)
    print(f'Last episode trajectory saved as {output_filename}')

# Updated run function to handle specific offline trial engagement
def run_agent_with_specific_offline_trials(alpha_td=0.5, alpha_sr=0.5, offline_samples=0, trials_per_episode=100, episodes=5, specific_trial=10, dynamic_lr=True,warmup_trials=0,rest_episode=0,reward_trials=0,epsilon=0.10,gamma=0.7):
    
    maze_size = 6
    reward_value = 1000
    punishment_value = -1
    reward_location = (3, 1)
    punishment_location = (1, 2)
    obstacles = []
    
    # Initialize the agent
    agent = SRDYNA(maze_size, gamma, epsilon, alpha_sr, alpha_td, offline_samples, 
                   reward_location, punishment_location, obstacles, reward_value, punishment_value, 
                   dynamic_lr, total_eps=episodes, rest_episode=rest_episode,specific_trial=specific_trial)
    
    # Warm-up Phase: Random Exploration -1000 trials default
    trajectory=agent.run_episode(0, warmup_trials,phase_general='warmup')

    for ep in range(episodes):
        trajectory=agent.run_episode(ep, trials_per_episode,phase_general='choice')
        # if ep==19:
        #     # Visualize the trajectory
        #     visualize_last_episode_trajectory(agent, trajectory, output_filename='last_episode.gif')

    return agent, agent.episode_goal_counts


# Configuration for multiple offline sample settings
specific_episodes = [1,2,3,4]  # Offline samples from 10 to 1000 in steps of 100
num_agents_per_setting = 100 # 10 agents per setting
trials_per_ep = 65
eps = 12
alpha_td = 0.2
alpha_sr = 0.2
epsilon=0.1
offline_samples=500


warmups=5000
gamma=0.99
# Collect data for all settings


data = []
for specific_episode in specific_episodes:
    print(specific_episode)
    for agent_id in range(num_agents_per_setting):
        agent, goal_counts = run_agent_with_specific_offline_trials(
            alpha_td=alpha_td,
            alpha_sr=alpha_sr,
            offline_samples=offline_samples,
            trials_per_episode=trials_per_ep,
            episodes=eps,
            specific_trial=2,
            dynamic_lr=False,
            warmup_trials=warmups,
            rest_episode=specific_episode,
            reward_trials=0,
            epsilon=epsilon,
            gamma=gamma
        )
        for episode_idx, count in enumerate(goal_counts):
            data.append({
                'OfflineStart': specific_episode,
                'Agent': f'Agent_{offline_samples}_{agent_id}',
                'Episode': episode_idx,
                'Goal Count': count
            })

```
